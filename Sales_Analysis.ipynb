## 3.2 Data Cleaning 

#### 1.Finding out all the NaN Values.

```python
nan_values = sales_2019[sales_2019.isna().any(axis=1)]
nan_values
```

##### Drop all NaN Values.

```python
sales_2019 = sales_2019.dropna(how='all')
```

```python
sales_2019.head()
```

```python
sales_2019.isna().sum()
```

##### All the NaN Values are removed.

#### 2.Duplicate Values

```python
sales_2019 = sales_2019.drop_duplicates()
```

```python
sales_2019[sales_2019.duplicated()]
```

```python
print('No. of Duplicate Values are:',sales_2019.duplicated().sum())
```

**Duplicate Values are removed.**

####  3.Order Date Column.

```python
sales_2019['Order Date']
```

**Lets extract Date,Month,Year,hour and  from Order Date**


```python
sales_2019[sales_2019['Order Date'].str[0:2] == 'Or']
```

- *We have an 'OR' value in Order Date column.*

```python
sales_2019 =sales_2019[sales_2019['Order Date'].str[0:2] != 'Or']
```

```python
sales_2019.head()
```

```python
sales_2019['Date'] =pd.to_datetime(sales_2019['Order Date']).dt.date
```

```python
sales_2019['Date'] = pd.to_datetime(sales_2019['Date'])
```

```python
sales_2019['Month'] = sales_2019['Order Date'].str[0:2]
sales_2019['Month'] = sales_2019['Month'].astype(int)
```

```python
sales_2019['Order Date'] = pd.to_datetime(sales_2019['Order Date'])
```

```python
sales_2019['Year'] =pd.to_datetime(sales_2019['Date']).dt.year
```

```python
sales_2019['Hour'] = pd.to_datetime(sales_2019['Order Date']).dt.hour
```

```python
sales_2019['Minute'] = pd.to_datetime(sales_2019['Order Date']).dt.minute
```

```python
sales_2019.head(5)
```

### 4.Data types

```python
sales_2019['Order ID'] = sales_2019['Order ID'].astype(int)
sales_2019['Quantity Ordered'] = sales_2019['Quantity Ordered'].astype(int)
sales_2019['Price Each'] = sales_2019['Price Each'].astype(float)
```

```python
sales_2019.dtypes
```

###  4.Add a Sales column

```python
sales_2019['Sales'] = sales_2019['Quantity Ordered'] * sales_2019['Price Each']
```

```python
sales_2019.head()
```

### 5.Add City column

```python
def get_city(address):
    return address.split(',')[1].strip(" ")
def get_state(address):
    return address.split(',')[2].split(" ")[1]
sales_2019['City'] = sales_2019['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
sales_2019.head()
```

```python
sales_2019.head()
```

```python
sales_2019.groupby('Month').sum()
```

```python
plt.figure(figsize=(16,8))
monthly_sales = sales_2019.groupby('Month').sum()['Sales']
month = np.arange(1,13)
plt.figure(figsize=(7,7))
plt.bar(month,monthly_sales)
plt.xticks(month)
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.show()
```

```python
# Best month for sale 
max_sale = sales_2019.groupby('Month').sum()['Sales'].max()
print("Maximum Sale is  in the month of December: ", max_sale)
```
